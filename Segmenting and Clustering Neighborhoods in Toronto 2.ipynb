<h2> import Data

```python
import pandas as pd
df=pd.read_csv('toronto.csv')
df.head()
```

```python
df_location=pd.read_csv('Geospatial_Coordinates.csv')
df_location.columns=['Postalcode','Latitude','Longitude']
df_location.head()
```

<h3> Merge Data

```python
df_final=pd.merge(df,df_location,how='inner',on='Postalcode')
df_final
```

```python
df_final.to_csv('d:/finaldf.csv')
```
